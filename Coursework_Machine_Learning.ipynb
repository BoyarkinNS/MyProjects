# 1. Описание данных

```python
#Считываем данные и описываем их.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('telecom_users.csv')
df.info()
```

Из полученных таблиц о данных можно предварительно заключить, что данные являются полными, пропущенные значения отсутствуют, также в количественных данных отсутствуют выбросы. Столбец TotalCharges имеет строковый тип. Вероятнее всего, из-за того, что какой-то клиент еще не оплатил услуги в связи с недавним подключением и в этом столбце присутствует либо пропуск, либо буквенное описание неоплаты. Заменим тип данной строки на float, предварительно заполнив ее числами в недостающих местах. Как можно заметить из результата работы функции info, большинство типов переменных в датасете - строки, определенный категориальный признак для каждого из пользователей. Необходимо перевести их в числовые переменнные для дальнейшей работы с даннными.

```python
#Удаляем ненужные столбцы, восстанавливаем нужный тип данных в датасете.
# Для построения графиков создадим датасет df_with_str с текстовым описанием величин. Для создания моделей ML заменим текстовые
# качественные величины на числовые по возрастанию, чтобы параметр 'yes' всегда принимал последнее значение (1 или 2) для
# избежания путаницы.

df = df.drop('customerID', axis=1)
df['TotalCharges'] = df['TotalCharges'].replace(' ', 0.0).astype('float')
df.drop('Unnamed: 0', axis=1, inplace=True)
titles_of_obj = df.select_dtypes('object').columns.to_list()
df_with_str = df.copy()
for title in titles_of_obj:
    df[title] = df[title].astype('category')
    df[title] = df[title].cat.codes
df[['tenure', 'MonthlyCharges','TotalCharges']].describe()
```

Как видно из полученных статистических характеристик, 50% клиентов пользуются услугами компании не более 29 месяцев, причем среднее число времени пользования составляет 32 месяца. Следовательно, есть небольшой перекос среди клиентов в сторону недавно подключившихся пользователей. Общая цена услуг за месяц варьируется от 18.25 у.е. до 118.75 у.е. Причем, медианное значение, которое не сильно отличается от среднего, составляет 64.8 у.е. Это говорит о том, что, в среднем, клиенты пользуются всеми услугами компании. Среднее и медианное значение TotalCharges сильно отличаются друг от друга. Это может значить, что существует большое количество недавно подключившихся клиентов, либо, что большинство клиентов пользуется базовыми и/или недорогими услугами. Так как анализ tenure показал, что недавно подключившихся клиентов не так много, можно сделать вывод, что чаще всего пользователи оплачивают базовые услуги и несколько недорогих дополнительных, судя по показателю MonthlyCharges.

# 2. Исследование зависимостей и формулирование гипотез


Необходимо сгруппировать факторы, которые могут влиять на уход пользователя.
Предположим, что первой группой факторов будет ценовая. Проверим, как сильно влияет на параметр churn такие параметры как общий счет за все время пользования, способ оплаты и платеж в месяц, тип договора. Также проверим, как влиял параметр времени пользования услугами компании.

```python
# Напишем функцию drawby для быстрого построения гистограмм.

def drawby(dataframe, titles):
    for title in titles:
        plt.figure(figsize=(10,5))
        sns.histplot(dataframe[df_with_str['Churn'] == 'Yes'][title], label='Churn == Yes', color='red')
        plt.legend()
        plt.figure(figsize=(10,5))
        sns.histplot(dataframe[df_with_str['Churn'] == 'No'][title], label='Churn == No', color='blue')
        plt.legend()
drawby(df_with_str, ['Contract', 'PaymentMethod', 'MonthlyCharges', 'TotalCharges','PaperlessBilling', 'tenure'])
```

По данным графикам видно, что чаще всего уходят те пользователи, которые не пользовались услугами компании долгое время, которые платили в месяц сумму выше среднего месячной оплаты и которые получили физический чек и электронный счет на оплату.
Также, если контракт был заключен на один или два года, вероятнее всего, пользователь не уйдет.

Сформулируем гипотезу о том, что именно из-за большого количества услуг, подключенных разом и/или неудовлетворенностью их качеством клиенты уходили. Проверим вторую категорию данных, к которой относятся услуги, которые пользователи подключили.

```python
#Воспользуемся функцией drawby для построения гистограмм.

drawby(df_with_str, ['PhoneService','MultipleLines','InternetService','OnlineSecurity',\
        'OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies'])
```

Данные гистограммы показывают то, что многим клиентам не понравился интернет-провайдер fiber optic и то, что большинство оставшихся клиентов пользуются дополнительными услугами, а ушедшие - нет. Проверим гипотезу о том, что пользователи уходили из-за большого количетсва услуг, подключенных одновременно. Добавим стоблец extras чтобы узнать, сколько услуг, помиио базовых интернета и связи, было у пользователя. Также, четко выделяется то, что, если у человека не подключен интернет от исследуемой компании, то он, вероятнее всего, останется.  
Следует выделить это в отдельный столбец.

Самыми влиятельными на уход клиента услугами были:
* Tech support
* Device protection
* Online backup
* Online security

```python
# Добавим параметр No_Int_Serv.

df['No_Int_Serv'] = df.apply(lambda x: 1 if x['StreamingMovies'] == 1 else 0, axis=1)
```

```python
# Посчитаем количество дополнительных услуг и построим их распределение.

def counter(x):
    
    count = 0
    extras = ['MultipleLines','OnlineSecurity',\
              'OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies']
    
    for extra in extras:
        if x[extra] == max(df[extra].unique()):
            count += 1
            
    return count

df['extras'] = df.apply(counter, axis=1)

plt.figure(figsize=(10,5))
sns.histplot(df[df['Churn'] == 1]['extras'], label='Churn == Yes', color='red')
plt.legend()
plt.figure(figsize=(10,5))
sns.histplot(df[df['Churn'] == 0]['extras'], label='Churn == No', color='blue')
plt.legend()
```

Гипотеза не подтвердилась, однако, стоит заметить, что чаще уходили те пользователи, у которых была подключена хотя бы одна дополнительная услуга, нежели те, у которых их не было.  

Проверим распределение ухода пользователей в зависимости от пола, возраста, семейного положения и наличия иждивенцев.

```python
#Воспользуемся функцией drawby для построения гистограмм.

drawby(df_with_str, ['gender', 'SeniorCitizen', 'Partner', 'Dependents'])
```

По полученным гистограммам видно, что если у клиента есть иждивенцы или он состоит в браке, то вероятность ухода такого клиента ниже, чем клиента с обратным семейным положением.  
Также, исходя из пропорциональных соотношений, пожилых людей среди ушедших было больше, чем пожилых людей среди оставшихся клиентов.

# 3. Построение моделей для прогнозирования оттока

Для прогнозирования ухода клиента построим две модели машинного обучения по методам бусинга и стэкинга. Базовые модели машинного обучения могут не подойти для этой задачи из-за большого количества переменных, влияющих на выбор того, уйдет клиент или нет.
Для нашего датасета значения целевой величины:
1 - клиент ушел
0 - клиент не ушел
Необходимо определиться с выбором метрики качества для настройки гиперпараметров моделей.
В данном случае, будут использоваться метрики accuracy_score, для общей качественной оценки модели и recall, т.к. нам важнее всего минимизировать количество ложных прогнозов о том, что клиент не ушел, то есть False Negative.

Для сглаживания данных возьмем натуральный логарифм от значения MonthlyCharges.

```python
from sklearn.model_selection import train_test_split
import xgboost as xgb

df_ML = df[['SeniorCitizen', 'Partner', 'Dependents', 'tenure', 'InternetService','TechSupport',\
            'DeviceProtection','OnlineBackup','OnlineSecurity','PaperlessBilling','PaymentMethod',\
            'MonthlyCharges','Contract','extras','No_Int_Serv', 'Churn']]
df_ML['MonthlyCharges'] = np.log(df_ML['MonthlyCharges'])

x_train, x_test, y_train, y_test = train_test_split(df_ML.drop('Churn', axis=1), df_ML.Churn, test_size=.3, random_state=17,\
                                                   stratify=df_ML.Churn)
dtrain = xgb.DMatrix(x_train, y_train)
dtest = xgb.DMatrix(x_test, y_test)

scale_pos_weight = 0.4* (len(y_train[y_train == 0]) / len(y_train[y_train == 1]))
params = {'objective': 'binary:logistic', 
          'max_depth': 3, 
          'eta': 0.02,
         'scale_pos_weight': scale_pos_weight}
num_rounds = 60
xgb_model = xgb.train(params, dtrain, num_rounds)
probability_matrix = xgb_model.predict(dtest)
result_matrix = [True if x>0.45 else False for x in probability_matrix]
```

Для того, чтобы обучить модель стэкинга, понадобится разбить весь датасет на train и test c пропорциями 0.6 и 0.4. Сначала на тестовом датасете обучим базовые модели (градиентный бусинг, RandomForestClassifier, SupportVector Classifier и логистическую регрессию), найдем при помощи кросс-валидации тестового датасета наилучшие параметры. Затем вновь обучим на датасете train базовые модели стэкинга, получим матрицу мета-признаков. На этой матрице обучим мета-алгоритм XGBClassifier и на той же кросс-валидации найдем наилучшие параметры. В конечном итоге, на датасете test обучим базовые модели стэкинга, получим матрицу мета-признаков и сделаем предсказания при помощи мета-алгоритма

```python
#Разобьем датасет df_ML на тренировочную и тестовую выборки, зададим параметры
# модели, обучим модель StackingClassifier и сделаем предсказание.
from sklearn.model_selection import (GridSearchCV,
                                     train_test_split,
                                     StratifiedKFold) 

from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier, StackingClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression

from xgboost import XGBClassifier


#  GridSearchCV(scoring='roc_auc', ....)  
gbc = GradientBoostingClassifier(learning_rate = 0.1)
rfc = RandomForestClassifier(n_estimators = 3, min_samples_leaf = 60)
svc = SVC(kernel = 'linear', C = 0.5)
lr = LogisticRegression(C = 0.7)
estimators = [('gbc', gbc), ('rfc', rfc), ('svc', svc), ('lr', lr)]
clf = StackingClassifier(estimators=estimators, final_estimator = XGBClassifier(n_estimators = 10, learning_rate = 0.1\
                                                                    ,scale_pos_weight =\
                                                                    1.2 * len(y_train[y_train == 0]) / len(y_train[y_train == 1])))
clf = clf.fit(x_train, y_train)
predictions = clf.predict_proba(x_test)

result_matrix_st = [True if x>0.55 else False for x in predictions[:,1]]
```

# 4. Сравнение качества моделей 

Для оценки качества моделей будем использовать целевую метрику roc-auc-score.  
Для интерпретации матрицы ошибок алгоритма и понимания того, насколько модель сбалансирована, будем использовать f1_score.

```python
# Распечатаем значение необходимых метрик.

from sklearn.metrics import confusion_matrix, roc_auc_score

result_matrix_array = np.array(result_matrix)
print(f'Матрица ошибок для алгоритма xgboost:\n  {confusion_matrix(y_test, result_matrix_array)}\n ')
print(f'roc_auc_score алгоритма xgboost: \n {roc_auc_score(y_test, result_matrix_array)}\n')
print(f'f1_score алгоритма xgboost: \n {f1_score(y_test, result_matrix_array)}\n')


result_matrix_st_array = np.array(result_matrix_st)
print(f'Матрица ошибок для алгоритма xgboost(стэкинг):\n  {confusion_matrix(y_test, result_matrix_st_array)}\n ')
print(f'roc_auc_score алгоритма xgboost(стэкинг): \n {roc_auc_score(y_test, result_matrix_st_array)}\n')
print(f'f1_score алгоритма xgboost(стэкинг): \n {f1_score(y_test, result_matrix_st_array)}\n')
```

Делаем вывод, что модель cтэкинга является лучшей для решения данной задачи. Ее общая точность выше, чем у модели бустинга, также метрика roc_auc_score показывает, что показатель качества наиболее важной величины для этого кейса выше у стэкинга. 
Также, в ходе работы были выявлены и созданы признаки, наиболее сильно отражающие факт того, уйдет клиент или нет.
