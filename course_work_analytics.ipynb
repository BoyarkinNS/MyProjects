Для того, чтобы проверить, что вы правильно соединили все датасеты в основной (без `students.csv`), запустите аналогичный приведенному ниже запрос. Выдача должна совпадать. Имейте в виду, что названия некоторых столбцов были переименованы.

```python
#Импортируем необходмые библиотеки и считываем датасеты.

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import datetime
import seaborn as sns
df_courses = pd.read_csv('Data/courses.csv', index_col=0)
df_course_contents = pd.read_csv('Data/course_contents.csv')
df_progress_phases = pd.read_csv('Data/progress_phases.csv')
df_progresses = pd.read_csv('Data/progresses.csv')
df_students = pd.read_csv('Data/students.csv', parse_dates=['birthday'], dayfirst=False)
```

```python
#Объединяем датасеты в один.

df_course_contents = df_course_contents.rename(columns={'course_id': 'id'})
df_courses = df_courses.merge(df_course_contents, how='right')
df_progress_phases = df_progress_phases.rename(columns={'progress_id':'id'})
df_progresses = df_progresses.merge(df_progress_phases, how='right')
df_courses = df_courses.rename(columns={'id':'course_id'})
df_courses = df_courses.merge(df_progresses, how='right')
```

```python
#Находим общее количество курсов в объединенном датасете 

print(f'Общее количество курсов в представленном датасете: {len(df_courses["title"].dropna().unique())}')
```

```python
#Группируем датасет по курсам, оставляем уникальные номера модулей. Затем, находим длину получившегося списка с модулями.
# Полученное значение является количеством модулей для каждого курса.

df_courses_with_modules = df_courses.groupby('title')['module_number'].unique().reset_index()
df_courses_with_modules['module_number'] = df_courses_with_modules['module_number'].apply(lambda x: len(x))
df_courses_with_modules = df_courses_with_modules.rename(columns={'module_number': 'amount_of_modules'})
print(f'Количество модулей на каждом курсе: \n\n {df_courses_with_modules}')
```

```python
#Группируем датасет по курсам и модулям, причем, для каджого модуля находим наибольшее значение номера урока.
# Данное значение является количеством уроков в каждом модуле.
df_courses_with_modules_with_lessons_max = df_courses.groupby(['title', 'module_number'])['lesson_number'].max().reset_index()
df_courses_with_modules_with_lessons_max = df_courses_with_modules_with_lessons_max\
.rename(columns={'lesson_number': 'amount_of_lessons'})
print(f'Количество уроков в каждом модуле на каждом курсе: \n\n {df_courses_with_modules_with_lessons_max}')
```

```python
#Создаем новый датасет, в котором группируем старый датасет по курсам и модулям и находим число уроков для каждого модуля.
# Затем, группируем полученный датасет только по курсам, находя медианное значение для числа уроков каждого из модулей.

df_courses_with_modules_with_lessons_med = df_courses.groupby(['title', 'module_number'])['lesson_number'].max().reset_index()\
.groupby('title')['lesson_number'].median().reset_index()

df_courses_with_modules_with_lessons_med = df_courses_with_modules_with_lessons_med\
.rename(columns={'lesson_number': 'med_amount_of_lessons_in_the_module'})
print(f'Медианное количество уроков в каждом модуле на каждом курсе: \n\n {df_courses_with_modules_with_lessons_med}')
```

```python
#Группируем датасет по курсам, находим уникальные id студентов, затем, находим число этих id для каждого полученного списка.df_students.info()

df_courses_with_students = df_courses.groupby('title')['student_id'].unique().reset_index()
df_courses_with_students['Amount_of_students'] = df_courses_with_students['student_id'].apply(lambda x: len(x))
df_courses_with_students.drop('student_id',axis=1, inplace=True)
print(f'Количество студентов на каждом курсе: \n\n {df_courses_with_students}')
df_students.info()
```

```python
#Cоздаем новый датафрейм, в котором нет пропусков в столбце 'birthday'
#Для нахождения возраста студентов проводим фильтрацию их даты рождения.
#Оставляем лишь студентов, чья дата рождения раньше 2005 и позже 1910
#Конвертируем даты рождения в datetime, удаляем неверно представленные данные, которые нельзя сконвертировать в datetime
#Находим разницу между текущей датой и датой рождения и находим статистические данные о возрасте студентов.

df_students_with_age = df_students[df_students['birthday'].notna()]

df_students_with_age.loc[:, 'birthday'] = df_students_with_age['birthday'].apply\
        (lambda x: x if (int(x[0:4]) < 2005) and (int(x[0:4]) > 1910) else None)
df_students_with_age = df_students_with_age[df_students_with_age['birthday'].notna()]
df_students_with_age.loc[:, 'birthday'] = df_students_with_age['birthday'].astype('datetime64[ns]')
df_students_with_age = df_students_with_age[df_students_with_age['birthday'].notna()]
today = datetime.datetime.now()
df_students_with_age['Age'] =  today - df_students_with_age['birthday']
df_students_with_age['Age'] = df_students_with_age['Age'].apply(lambda x : int(int(x.days) / 365))
```

```python
#Распечатываем статистические данные о возрасте студентов.
print(f'Минимальный возраст студента: {df_students_with_age.Age.min()} \n \
Средний возраст студента: {df_students_with_age.Age.mean()} \n \
Максимальный возраст студента: {df_students_with_age.Age.max()} \n \
Медианный возраст студента: {df_students_with_age.Age.median()}')
```

```python
#Для каждого студента в исходном датафрейме добавляем его возраст из датафрейма df_students_with_age если он известен
age_ids = list(df_students_with_age.id)
df_courses['Age'] = df_courses['student_id'].apply\
(lambda x: int(df_students_with_age[df_students_with_age['id'] == x]['Age']) if x in age_ids else None)
```

```python
#Создаем новый датафрейм df_courses_age в котором стоблец возраста студенов не содержит пропусков
# По курсам находим все уникальные id для каждого курса, заменяем id на возраст стуендта из 
# колонки Age датасета df_students_with_age
df_courses_age = df_courses[df_courses['Age'].notna()]
df_courses_age = df_courses_age.groupby('title')['student_id'].unique().reset_index()
df_courses_age['Age'] = df_courses_age['student_id'].apply\
(lambda x: [int(df_students_with_age[df_students_with_age['id'] == student_id]['Age']) for student_id in x])
df_courses_age.drop('student_id', axis=1, inplace=True)
```

```python
# В датасете df_courses_age создаем столбцы с статистическими парамертами возраста студентов по каждому курсу и заполнеяем их.
import statistics
df_courses_age['Age_max'] = df_courses_age.Age.apply(lambda x: max(x))
df_courses_age['Age_min'] = df_courses_age.Age.apply(lambda x: min(x))
df_courses_age['Age_avr'] = df_courses_age.Age.apply(lambda x: sum(x)/len(x))
df_courses_age['Age_median'] = df_courses_age.Age.apply(lambda x: statistics.median(x))
print('Минимальный, максимальный, средний, медианный возраст студентов на каждом курсе:\n')
df_courses_age.drop("Age", axis=1)
```

```python
# Сортируем значения датафрейма по колонке числа студентов, добавляем параметр область из исходного датасета
# Строим барплот, указав в качестве параметра hue переменную области курса field.
# Убираем рамки и регулируем ширину столбцов

df_courses_with_students = df_courses_with_students.sort_values('Amount_of_students', ascending=False)
df_courses_with_students['field'] = df_courses_with_students\
    ['title'].apply(lambda x: df_courses[df_courses['title'] == x]['field'].mode()[0])
plt.figure(figsize=(15,10))
ax = sns.barplot(y=df_courses_with_students.title, \
    x=df_courses_with_students.Amount_of_students, palette='rocket', alpha=0.9, \
    hue=df_courses_with_students.field,  linewidth=1000)
for patch in ax.patches:
    patch.set_height(0.6)

plt.axvline(df_courses_with_students['Amount_of_students'].median(), color='blue', label='Медианное количество студентов',\
           alpha=0.5)
for key, spine in ax.spines.items():
    spine.set_visible(False)
plt.legend()
plt.show()
```

```python
# Исходным датафреймом для данного задания будет являться df_for_third_module, созданный во втором модуле при выполнении
# второго пункта. Он отличается тем, что пустые значения в колонке finish_date удаляются, а не заполняются максимальной датой,
# как во втором модуле.
# Для того, чтобы найти статистические показателя прохождения модулей, создадим пустой датафрейм df_completing_of_mods
#с названиями курсов,  их модулями, и столбцом time отражающим время прохождения модуля.
# Затем, создадим список студентов в переменной list_of_students. Для каждого из студентов будем выделять нужную область в
# датафрейме содержащую домашнее задание и, проходя по срезанному датафрейму по названиям курсов, которые проходит студент,
# добавим время проходжения модуля в датафрейм с курсами и их модулями. Также учтем, что в модуле может содержаться более 1 дз
# поэтому будем находить время как разность между finish_date последнего урока для данного модуля и start_date первого урока.
# Для отображения данных создадим датафрейм df_completing_of_mods_to_show содержащий только название курса,
# модуль и статистические данные, а не все записи по прохождениям курсов и заполним его.
df_completing_of_mods = pd.DataFrame(columns=['title','module_number', 'time'])
list_of_students = list(df_for_third_module.student_id.unique())
df_completing_of_mods_to_show = df_completing_of_mods.copy()
df_for_third_module['finish_date'] = df_for_third_module['finish_date'].astype('datetime64[ns]')
df_for_third_module['start_date'] = df_for_third_module['start_date'].astype('datetime64[ns]')
for student_id in list_of_students:
    
    slice_for_one_student = df_for_third_module[(df_for_third_module['is_homework'] == True) \
                                              & (df_for_third_module['student_id'] == student_id) \
                                                & (df_for_third_module['start_date'] < df_for_third_module['finish_date'])]
    list_of_students_titles = list(slice_for_one_student.title.unique())
    if list_of_students_titles is None:
        continue
    else:
    
        for title in list_of_students_titles:
        
            slice_for_title = slice_for_one_student[slice_for_one_student['title'] == title]
            modules_in_slice = list(slice_for_title.module_number.unique())
        
            for module_number in modules_in_slice:
                slice_with_one_module = slice_for_title[slice_for_title['module_number'] == module_number]
                slice_with_one_module = slice_with_one_module.sort_values('lesson_number')
                value_to_append = (slice_with_one_module.iloc[-1]['finish_date']\
                                   - slice_with_one_module.iloc[0]['start_date']).days
                new_row = {'title':title, 'module_number':module_number, 'time':value_to_append}
                df_completing_of_mods = df_completing_of_mods.append(new_row, ignore_index=True)
df_completing_of_mods['time'] = df_completing_of_mods['time'].astype('float')
df_completing_of_mods_to_show = df_completing_of_mods.groupby(['title', 'module_number'])\
.mean().reset_index()[['title', 'module_number']]
df_completing_of_mods_to_show['min'] = df_completing_of_mods.groupby(['title', 'module_number']).min().reset_index()['time']
df_completing_of_mods_to_show['max'] = df_completing_of_mods.groupby(['title', 'module_number']).max().reset_index()['time']
df_completing_of_mods_to_show['mean'] = df_completing_of_mods.groupby(['title', 'module_number']).mean().reset_index()['time']
df_completing_of_mods_to_show['median'] = df_completing_of_mods.groupby(['title', 'module_number'])\
.median().reset_index()['time']
print('Статистические данные о времени прохождения модулей: \n')
df_completing_of_mods_to_show
```

```python
#Найдем первый и последний урок для каждого модуля в каждом курсе, создадим датасеты df_started и df_finished, содержащие только
# первые и последние уроки с дз соответственно, найдем количество учеников, начавших и закончивших данные уроки.
# Уберем из датасета df_started первые модули с дз в каждом курсе, т.к. конверсия считается со второго.
# При помощи функции convertion будем делить количество учеников
# приступивших к дз в текущем модуле на количетство учеников, закончивших предыдущий.
df_for_forth_module['token'] = 1
df_last_lessons = df_for_forth_module[(df_for_forth_module['is_homework'] == True)]
df_last_lessons = df_last_lessons.groupby(['title', 'module_number'])['lesson_number'].last().reset_index()
df_first_lessons = df_for_forth_module[(df_for_forth_module['is_homework'] == True)]
df_first_lessons = df_first_lessons.groupby(['title', 'module_number'])['lesson_number'].first().reset_index()
df_started = df_for_forth_module.copy()
df_started = df_started.merge(df_first_lessons, how='right')
df_finished = df_for_forth_module[df_for_forth_module['status'] == 'done'].copy()
df_finished = df_finished.merge(df_last_lessons, how='right')
df_started = df_started.groupby(['title', 'module_number'])['token'].sum().reset_index()
df_finished = df_finished.groupby(['title', 'module_number'])['token'].sum().reset_index()
df_started = df_started.groupby('title').apply(lambda group: group[1:])
df_started.index = df_started.index.droplevel()

def convertion(x):
    current_module = int(x['module_number'])
    title_name = str(x['title'])
    ind = df_finished[(df_finished['title'] == title_name) & (df_finished['module_number'] == current_module)].index[0]
    
    if (x['token'] == 0) | (df_finished.loc[ind-1]['token'] == 0):
        return 0
    
    else:
        conv = (x['token'] / (df_finished.loc[ind-1]['token']))
        return conv
    
df_started['conv'] = df_started.apply(convertion, axis=1)
```

```python
#Посторим значения конверсий на одном графике.

plt.figure(figsize = (15,70))
sns.barplot(y = df_started.module_number, x = df_started.conv, orient='h', hue = df_started.title, palette='rocket')
```
